```python
from openmm.app import *
from openmm import *
from openmm.unit import *
from sys import stdout
```

```python
pdb = PDBFile("6zgj_cleaned.pdb")
```

```python
forcefield = ForceField("amber14-all.xml", "amber14/tip3pfb.xml")
```

```python
system = forcefield.createSystem(pdb.topology, nonbondedMethod=CutoffNonPeriodic, nonbondedCutoff=1*nanometer, constraints=HBonds)
```

```python
integrator = LangevinMiddleIntegrator(300*kelvin, 1/picosecond, 0.004*picosecond)
```

```python
simulation = Simulation(pdb.topology, system, integrator)
```

```python
simulation.context.setPositions(pdb.positions)
```

```python
simulation.minimizeEnergy()
```

```python
simulation.reporters.append(PDBReporter("output.pdb", 1000))
```

```python
simulation.reporters.append(StateDataReporter(stdout, 1000, step=True, potentialEnergy=True, temperature=True))
```

```python
simulation.step(10000)
```
